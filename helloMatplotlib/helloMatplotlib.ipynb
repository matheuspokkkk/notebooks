```python
import matplotlib.pyplot
#importa o módulo pyplot da lib matplotlib
```

```python
matplotlib.pyplot.plot([1, 2, 3, 4, 5], [5, 4, 3, 2, 1], 'ro')
#cria um plot 2D, de maneira que as coordenadas de cada um estão listadas na primeira lista (x) e na segunda (y)
#os pontos, são portanto: (1,5), (2,4), ... , (5, 1) 
#por fim, diz que os pontos serão representados como red circles (bolinhas vermelhas)

matplotlib.pyplot.ylabel("descrição do eixo y")
matplotlib.pyplot.xlabel("descrição do eixo x")
#põe label nos eixos

matplotlib.pyplot.axis([0, 10, 0, 7])
#diz que o grafico mostrará os valores de x de 0 até 10 e os valores de y de 0 até 7

matplotlib.pyplot.show()
#imprime o gráfico na tela
```

```python
import numpy
#importa a lib numpy
```

```python
vector = numpy.array([0, 1, 2, 3, 4, 5, 6])
#cria um vetor usando o método array do numpy, e armazena os valores de 0 à 6 neste vetor

matplotlib.pyplot.plot(vector, vector, 'r--', vector, vector**2, 'b--')
#aqui eu estou plotando duas representações diferentes em um mesmo gráfico
#o primeiro plot tem como os valores de x e y os valores do vetor
#o primeiro é representado por red-- (tracinhos vermelhos)
#o segundo tem como valores de x os valores do vetor, porém y recebe os valores do vetor ao quadrado
#o segundo é representado por tracinhos azuis
#um gráfico é linear e outro é quadrático
#posso imprimir diversos plots juntos em um único gráfico

matplotlib.pyplot.show()
#imprime o gráfico na tela
```

#### Plotando vários gráficos em uma única imagem (subplots):

```python
bigPicture, smallPics = matplotlib.pyplot.subplots(nrows = 2, ncols = 2)
#bigPicture == variável que cuida da "imagem geral", que conterá todos os gráficos.
#smallPicss == variável que armazena a lista de gráficos.
#para mexer na imagem geral, mexo na variável bigPicture, e para mexer em gráficos individuais, na smallPics
#no fim, digo que a imagem terá dois gráficos em cada coluna e também dois gráficos em cada linha

vector = numpy.array([0, 1, 2, 3, 4, 5, 6])
#novamente criando um vetor

smallPics[0][0].hist(vector, vector)
smallPics[0][1].plot(vector, vector)
smallPics[1][0].plot(vector, vector, 'ro')
smallPics[1][1].scatter(vector, vector)
#montando o gráfico da linha 0, coluna 0.. será um histograma
#montando o gráfico da linha 0, coluna 1, será um plot normal
#montando o gráfico da linha 1, coluna 0: plot normal cujos pontos são bolinhas vermelhas
#montando linha 1, coluna 1: gráfico de dispersão

bigPicture.suptitle("Titulo da Imagem Geral")
#atribui um titulo à imagem que contem os gráficos

matplotlib.pyplot.show()
#imprime o gráfico na tela

matplotlib.pyplot.savefig("nomeDoArquivo.png")
#salva a imagem com os plots no diretorio que que o código for executado, com o nome do arquivo estipulado
```

```python
import pandas
#importa a lib pandas
```

```python
irisDataFrame = pandas.read_csv("iris.csv", header = None)
#cria um data frame com o conteúdo do arquivo iris.csv, e sinaliza que este arquivo não possui header
```

_**Data Frame** é um tipo de estrutura de dados que a lib nos permite trabalhar._

*Possuem duas dimensões e portanto são especialmente úteis para dados estruturados em **tabelas**.*

*São objetos onde podem ser armazenados diversos outros tipos de dados, como vetores, listas ou outros data frames*

```python
irisDataFrame
#retorna o conteúdo do data frame iris
```

```python
irisDataFrame.columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Name"]
#atribui um header ao data frame caso ele não possua um
#se já possuir um header, ele o altera para este que está sendo passado
#a alteração é feita pela troca da primeira linha do data frame pelo conteúdo passado neste comando
```

```python
irisDataFrame
#retornando o conteúdo do data frame iris, que agora possui header
```

```python
irisDataFrame.head(10)
#método para retornar apenas as dez primeiras linhas do data frame
```

```python
irisDataFrame[:10]
#outra maneira de retornar as dez primeiras linhas do data frame
```

```python
irisDataFrame.tail(10)
#retorna as dez ultimas linhas
```

```python
irisDataFrame.shape
#retorna o número de linhas e o número de colunas do data frame
```

```python
irisDataFrame[irisDataFrame["Name"] == "Virginica"].shape
#retorna o número de linhas e colunas cujos valores de name são iguais à virginica
```

```python
irisDataFrame.count()
#retorna o número de linhas que cada coluna possui
#é especialmente útil para quando o csv está mais bagunçado, como quando possui células vazias por exemplo
```

```python
irisDataFrame["Name"].count()
#retorna o número de linhas existentes na coluna "name"
```

```python
irisDataFrame["Name"].value_counts()
#retorna a frequência de cada valor presente na coluna "name"
```

```python
irisDataFrame.dtypes
#retorna os tipos de dados dos valores presentes nas colunas do data frame
#strings são representadas como objetos
```

```python
irisDataFrame.describe()
#método que retorna, para cada coluna:
#o número de linhas; a média dos valores; o desvio padrão; o valor mínimo; o valor máximo; e os quartis.
```
